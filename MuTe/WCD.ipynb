### GPS time stamp analysis

To verify the GPS time stamp the ./lago -g was used in order to record time data directly fron the GPS

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.mplot3d import Axes3D
import scipy
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import scipy.ndimage as ndimage
from datetime import datetime
import pandas as pd

# This ensures visualizations are plotted inside the notebook
%matplotlib inline
```

```python
f = 'WCD/Date2.dat'    # Path to file containing the table

df = pd.read_csv(f, header=None, skiprows=[0, 1, 2], sep=' ')
df = df[[0]]
df.columns = ['time']
df.head()
```

```python
time = [datetime.strptime(i, "%m/%d/%YT%H:%M:%SZ")for i in df['time']]
```

```python
plt.figure(figsize=(16, 8))
plt.plot(time,'or',alpha=0.5)
plt.xlabel('Sample', fontsize=25)
plt.ylabel('Date', fontsize=25)
plt.grid()
plt.show()
```
