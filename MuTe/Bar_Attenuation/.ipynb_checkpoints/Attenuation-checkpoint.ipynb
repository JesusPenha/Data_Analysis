### Scitillator attenuation

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.mplot3d import Axes3D
import scipy
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import scipy.ndimage as ndimage

# This ensures visualizations are plotted inside the notebook
%matplotlib inline
```

```python
data = np.loadtxt('data/115cm.txt')
pulses = -(data.T)
pulses = pulses[200:250,0:100]
```

```python
N = len(pulses.T)
max_val = np.zeros(N)

for i in range(N):
    max_val[i] = np.max(pulses[:,i])
mean = np.mean(max_val)
```

```python
fig = plt.figure(figsize=(10, 5))
plt.plot(pulses*1000, 'r', alpha=0.1)
plt.xlabel('Time (* 8ns)')
plt.ylabel('Amplitude (mV)')
plt.grid()
plt.show()
```

```python
step = np.linspace(5,115,23)
M = len(step)

mean = np.zeros(M)
std = np.zeros(M)
```

```python
for j in range (M):
    
    name = str(int(step[j])) + "cm.txt"
    data = np.loadtxt(name)
    pulses = -(data.T)
    pulses = pulses[200:250,:]
    
    
    N = len(pulses.T)
    max_val = np.zeros(N)

    for i in range(N):
        max_val[i] = np.max(pulses[:,i])
        
    mean[j] = np.mean(max_val)
    std[j] = np.std(max_val)
```

```python
xerr = step - step
yerr = std
```

```python
fig = plt.figure(figsize=(6, 5))
plt.plot(step, mean, 'o')
plt.errorbar(step, mean, yerr, xerr, fmt='.', linewidth=2.0)
plt.xlabel('Distance [cm]', fontsize=20)
plt.ylabel('Amplitude [V]', fontsize=20)
plt.grid()
plt.show()
```

### Attenuation function fit

```python
from scipy.optimize import curve_fit

# this is the function we want to fit to our data
def func(x, a, b):
    'nonlinear function in a and b to fit to data'
    return a * np.exp(-b*x)


initial_guess = [1, 0.2]
pars, pcov = curve_fit(func, step, mean, p0=initial_guess)

xfit = np.linspace(0,120,101)
yfit = func(xfit, pars[0], pars[1])

print "Attenuation = " + str(np.min(yfit)/np.max(yfit))
```

```python
fig = plt.figure(figsize=(7, 5))
plt.errorbar(step, mean/np.max(yfit), yerr, xerr, fmt='.g', linewidth=2.0)
plt.plot(step, mean/np.max(yfit), 'ob', label="data")
plt.plot(xfit,yfit/np.max(yfit),'--k', label="fit")
plt.xlabel('Distance [cm]', fontsize=20)
plt.ylabel('Amplitude ratio [%]', fontsize=20)
plt.legend(loc=0)
plt.grid()
plt.show()
```
